**3. Compare and contrast mutable and immutable objects in Python with examples?**


- **Mutable Objects**


*   **Definition :** Mutable objects can be modified after they are created. This means you can change their value or content without creating a new object.

*   **Examples :**
  
*   Lists (list)
*   Dictionaries (dict)
*   Sets (set)


```python
my_list = [1, 2, 3]
my_list[0] = 4 # Modifying the list in-place
print(my_list) # Output : [4, 2, 3]
```

**Immutable Objects**

*  **Defination** **:** Immutable objects cannot be modified after they are created. If you try to change their value, a new object is created instead.

*    **Example :**
*   Numbers (int, float, complex)
Strings (str)
*   Tuples (tuple)
*   Frozen sets (frozenset)

```python
my_string = "hello"
my_string += "world"  # Creates a new string Object
print(my_string)   # output : " hello world"
```

**In Summmary :**


*   Mutable objects allow in-place modifications, while immutable objects require creating new objects for any changes.

*  Understanding this difference is crucial for writing efficient and bug-free Python code

1.   **Arithmetic Operators**: Used For Mathematical Calculation.
  


```python
x = 10
y = 5
print(x + y)  # Addition
print(x - y)  # Subtraction
print(x * y)  # Multiplication
print(x / y)  # Division
print(x % y)  # Modulus (Remainder)
print(x ** y) # Exponentiation
print(x // y) # Floor Division
```

**2. Comparison Operaors :** Compare Values and Return true or false.

```python
a = 5
b = 10
print(a == b) #Output : False (Equal to)
Print(a != b) #Output : True (Not Equal to)
print(a < b)  #Output : True (Less Than)
print(a > b)  #Output : False (Greater Than)
print(a <= b) #Output : True (Less Than or Equal to)
print(a >= b) #Output : False (Greater Than or Equal to)
```

**3. Logical Operators :**Combine Boolean expressions (true/false).

```python
p = True
q = False
print(p and q) #Output : False (Logical AND)
print(p or q)  #Output : True (Logical OR)
print(not p)   #Output : False (Logical NOT)
```

**4. Assignment Operators :**Assign values to variables.

```python
x = 10
x += 5 # Equivalent to x = x + 5
print(x) #Output : 15
x -= 3 # Equivalent to x = x - 3
print(x) #Output : 12
x *= 2 # Equivalent to x = x * 2
print(x) #Output : 24
x /= 4 # Equivalent to x = x / 4
print(x) #Output : 6.0
x %= 3 # Equivalent to x = x % 3
print(x) #Output : 0.0
x **= 2 # Equivalent to x = x ** 2
print(x) #Output :
x //= 2 # Equivalent to x = x // 2
print(x) #Output : 0.0
```

**5. Bitwise Operators :** Operators on individual bits of data (less common).

```python
a = 5 # Binary: 00000101
b = 3 # Binary: 00000011
print(a & b)  # Bitwise AND: 0
print(a | b)  # Bitwise OR: 7
print(a ^ b)  # Bitwise XOR: 6
print(~a)     # Bitwise NOT:
print(a << 2) # Left shift: 20
print(a >> 2) # Right shift: 1
```

**6. Membership Operators :** Check if a value is present in a sequence.

```python
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # Output: True
print(6 in my_list)  # Output: False
print(3 not in my_list)  # Output: False
print(6 not in my_list)  # Output: True
```

**5. Explain the concept of type casting in Python with examples?**


**Type Casting in Python**

Type casting, also known as type conversion, is the process of changing the data type of a variable from one type to another. Python provides several built-in functions to perform type casting. Here are some common type casting functions:

- **int()**: Converts a value to an integer.
- f**loat()**: Converts a value to a floating-point number.
- **str**(): Converts a value to a string.
- **bool**(): Converts a value to a boolean.
- **list**(): Converts a value to a list.
- **tuple**(): Converts a value to a tuple.
- **set**(): Converts a value to a set.
- **dict**(): Converts a value to a dictionary.


**Examples :**
Here are some examples of how to use type casting function in Python.

```python
# Convert a string to an integer
x ="10"
y = int(x)
print(y) # Output: 10

# Convert an Integer to a string
x = 10
y = str(x)
print(y)  # Output : "10"


# Convert a string to a float
x = "3.14"
y = float(x)
print(y)   # output : 3.14

# Convert a float to an integer
x = 3.14
y = int(x)
print(y)  # Output: 3

# Convert a list to a tuple
x = [1, 2, 3]
y = tuple(x)
print(y)  # Output: (1, 2, 3)
```

Conditional statements allow you to control the flow of your program based on certain conditions. The most common conditional statements in Python are:

1. **if statement**: Executes a block of code if a condition is True.

2. **elif statement**: Executes a block of code if the previous conditions were False and the current condition is True.

3. **else statement**: Executes a block of code if all previous conditions were False.

**Syntax**

```python
if condition1:
  # Code to execute if condition1 is True
elif condition2:
  # Code to execute if condition1 is False and condition2 is True
else:
  # Code to execute if all previous conditions are False
```

```python
# Example 1: Check if a number is positive, negative, or zero
number = 10

if number > 0:
  print("The number is positive.")
elif number < 0:
  print("The number is negative.")
else:
  print("The number is zero.")

# Example 2: Check if a student is eligible for a scholarship
score = 85
attendance = 90

if score >= 90 and attendance >= 95:
  print("The student is eligible for a scholarship.")
else:
  print("The student is not eligible for a scholarship.")
```

**7.Describe the different types of loops in Python and their use cases with examples?**


**Types of Loops in Python**

Python provides two primary types of loops:

1. **for loop**: Used to iterate over a sequence (such as a list, tuple, string, or range) or other iterable objects.

2.**while loop**: Used to repeatedly execute a block of code as long as a condition is True.

**Use Cases and Examples**

Here's a breakdown of their use cases and examples:

**for loop**

- **Use Case**: When you know the number of iterations in advance or need to iterate over a specific sequence.

- **Example :**

```python
# Print numbers from 0 to 4
for i in range(5):
  print(i)

# Iterate over a list of fruits
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)
```

**while loop**

- **Use Case:** When you need to repeat a block of code until a specific condition becomes False.
- **Example**:

```python
# Print numbers from 0 to 4 using a while loop
count = 0
while count < 5:
  print(count)
  count += 1

# Keep asking for input until the user enters "quit"
user_input = ""
while user_input != "quit":
  user_input = input("Enter a command (or 'quit' to exit): ")
  print("You entered:", user_input)
```

**Loop Control Statements**

You can use loop control statements to modify the behavior of loops:

- **break:** Terminates the loop prematurely.

```python
for i in range(10):
  if i == 5:
    break  # Exit the loop when i is 5
  print(i)
```

- **continue**: Skips the current iteration and proceeds to the next.

```python
for i in range(10):
  if i % 2 == 0:
    continue  # Skip even numbers
  print(i)
```

**Choosing the Right Loop**

- Use a **for** loop when you know the number of iterations or need to iterate over a sequence.
- Use a **while** loop when you need to repeat code based on a condition that may change during execution
